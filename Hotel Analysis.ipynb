```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('./zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

# Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
# Cleaning rate column 

df['rate'].unique()
```

```python
# Removing "NEW" , "-" and "/5" from Rate Column

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

# Filling Null values with mean in Rate

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# Listed in(city) and location, both are there, so we will keep only one.

df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['Cost2plates'].unique()
```

```python
# Removing , from Cost2Plates Column

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python

# Cleaning Rest_type column

rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
# Making Rest Types less than 1000 in frequency as others

def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
        
```

```python
# cleaning Location column

location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
# cleaning Cuisines column

cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

# Data Visualization

```python

# Counting plot of various locations :-

plt.figure(figsize=(16, 10))
ax = sns.countplot(data=df, x='location')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Count of Restaurants')
plt.title('Count of Restaurants in Different Locations')
plt.show()
```

```python
# Visualizing Online order

plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='online_order', palette='inferno')
plt.xlabel('Online Order Availability')
plt.ylabel('Count')
plt.title('Count of Restaurants Offering Online Order')
plt.show()
```

```python
# Visualizing book Table

plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='book_table', palette='rainbow')
plt.xlabel('Book Table Availability')
plt.ylabel('Count')
plt.title('Count of Restaurants Offering Table Booking')
plt.show()
```

```python
# Visualizing online order and rate

plt.figure(figsize=(6, 6))
sns.boxplot(x='online_order', y='rate', data=df, order=['Yes', 'No'])
plt.xlabel('Online Order Availability')
plt.ylabel('Rate')
plt.title('Online Order vs Rate')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='book_table', y='rate', data=df, order=['Yes', 'No'])
plt.xlabel('Book Table Availability')
plt.ylabel('Rate')
plt.title('Book Table vs Rate')
plt.show()
```
